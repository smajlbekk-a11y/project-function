```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv('/content/results/combined_history.csv')

plt.figure(figsize=(8,5))
sns.lineplot(data=data, x='epoch', y='val_acc', hue='activation', marker='o')
plt.title("Validation Accuracy –±–æ–π—ã–Ω—à–∞ —Å–∞–ª—ã—Å—Ç—ã—Ä—É")
plt.xlabel("Epoch")
plt.ylabel("Validation Accuracy")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.lineplot(data=data, x='epoch', y='train_loss', hue='activation', marker='o')
plt.title("Train Loss –¥–∏–Ω–∞–º–∏–∫–∞—Å—ã")
plt.xlabel("Epoch")
plt.ylabel("Train Loss")
plt.grid()
plt.show()
```

```python
summary = data.groupby('activation')[['val_acc', 'train_loss', 'epoch_time_sec', 'grad_norm']].mean().reset_index()

summary_melted = summary.melt(id_vars='activation', var_name='metric', value_name='value')

plt.figure(figsize=(8,5))
sns.barplot(data=summary_melted, x='activation', y='value', hue='metric')
plt.title("”ò—Ä —Ñ—É–Ω–∫—Ü–∏—è–Ω—ã“£ –æ—Ä—Ç–∞—à–∞ –Ω”ô—Ç–∏–∂–µ–ª–µ—Ä—ñ")
plt.xlabel("–ê–∫—Ç–∏–≤–∞—Ü–∏—è —Ç“Ø—Ä—ñ")
plt.ylabel("–û—Ä—Ç–∞—à–∞ –º”ô–Ω")
plt.grid()
plt.show()
```

```python
import numpy as np
from math import pi

categories = ['val_acc', 'train_loss', 'grad_norm', 'epoch_time_sec']
N = len(categories)

summary = data.groupby('activation')[categories].mean()

plt.figure(figsize=(7,7))
for i, (act, row) in enumerate(summary.iterrows()):
    values = row.tolist()
    values += values[:1]
    angles = [n / float(N) * 2 * pi for n in range(N)]
    angles += angles[:1]
    plt.polar(angles, values, label=act)
plt.title("Activation —Ñ—É–Ω–∫—Ü–∏—è–ª–∞—Ä—ã–Ω—ã“£ –∂–∞–ª–ø—ã –∫”©—Ä—Å–µ—Ç–∫—ñ—à—Ç–µ—Ä—ñ")
plt.legend(loc='upper right')
plt.show()
```

```python
summary = data.groupby('activation').agg({
    'val_acc': ['mean', 'max'],
    'train_loss': ['mean', 'min'],
    'epoch_time_sec': 'mean',
    'grad_norm': 'mean'
}).round(4)

print(summary)
```

```python
import seaborn as sns
corr = data[['train_loss', 'val_acc', 'grad_norm', 'epoch_time_sec']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("–ú–µ—Ç—Ä–∏–∫–∞–ª–∞—Ä –∞—Ä–∞—Å—ã–Ω–¥–∞“ì—ã –∫–æ—Ä—Ä–µ–ª—è—Ü–∏—è")
plt.show()
```

```python
data['efficiency'] = data['val_acc'] / data['epoch_time_sec']
sns.barplot(data=data, x='activation', y='efficiency', estimator='mean')
plt.title("Efficiency (Accuracy per second)")
plt.show()
```

```python
import os
import pandas as pd
import subprocess

activations = ["sigmoid", "tanh", "relu"]
learning_rates = [0.1, 0.01, 0.001, 0.0001]

results_dir = "/content/results"
os.makedirs(results_dir, exist_ok=True)

combined_data = []

# üîÅ ”ò—Ä activation –∂”ô–Ω–µ ”ô—Ä learning rate –∫–æ–º–±–∏–Ω–∞—Ü–∏—è—Å—ã–Ω —Ç–µ–∫—Å–µ—Ä—É
for act in activations:
    for lr in learning_rates:
        print(f"\nüöÄ Training started with activation = {act.upper()}, learning_rate = {lr}\n")

        # –ê—Ä–≥—É–º–µ–Ω—Ç—Ç–µ—Ä–º–µ–Ω train.py —ñ—Å–∫–µ “õ–æ—Å—É
        subprocess.run([
            "python", "src/train.py",
            "--activation", act,
            "--lr", str(lr)   # ‚úÖ —Ç“Ø–∑–µ—Ç—ñ–ª–¥—ñ
        ])

        # –ù”ô—Ç–∏–∂–µ —Ñ–∞–π–ª –∞—Ç–∞—É—ã (–º—ã—Å–∞–ª—ã: results/sigmoid_lr0.01_history.csv)
        history_path = os.path.join(results_dir, f"{act}_lr{lr}_history.csv")
        if os.path.exists(history_path):
            df = pd.read_csv(history_path)
            df["activation"] = act
            df["learning_rate"] = lr
            combined_data.append(df)

# üîó –ë–∞—Ä–ª—ã“õ –Ω”ô—Ç–∏–∂–µ–ª–µ—Ä–¥—ñ –±—ñ—Ä—ñ–∫—Ç—ñ—Ä—É
if combined_data:
    combined_df = pd.concat(combined_data, ignore_index=True)
    combined_path = os.path.join(results_dir, "combined_lr_history.csv")
    combined_df.to_csv(combined_path, index=False)
    print(f"\n‚úÖ All experiments saved to {combined_path}")
else:
    print("‚ö†Ô∏è No result files found. Check your train.py output paths.")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv('/content/results/combined_lr_history.csv')

plt.figure(figsize=(10, 6))
palette = sns.color_palette("husl", len(data['learning_rate'].unique()))  # –∞—à—ã“õ —Ç“Ø—Å—Ç–µ—Ä

sns.lineplot(
    data=data,
    x='epoch',
    y='val_acc',
    hue='learning_rate',
    style='activation',
    palette=palette,
    linewidth=2.5
)

plt.title("Validation Accuracy vs Learning Rate (”ô—Ä Activation “Ø—à—ñ–Ω)", fontsize=14)
plt.xlabel("Epoch")
plt.ylabel("Validation Accuracy")
plt.grid(True, linestyle="--", alpha=0.5)
plt.legend(title="Learning Rate", fontsize=9)
plt.show()
```

```python
data['efficiency'] = data['val_acc'] / data['epoch_time_sec']
sns.barplot(data=data, x='learning_rate', y='efficiency', hue='activation')
plt.title("Efficiency (Accuracy per second) by Learning Rate and Activation")
plt.show()
```
